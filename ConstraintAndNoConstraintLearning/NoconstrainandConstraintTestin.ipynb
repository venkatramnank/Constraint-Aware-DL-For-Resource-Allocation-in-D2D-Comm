## Knapsack

```python
import numpy as np
import pandas as pd
import math
```

```python

def brute_force_knapsack(x_weights, x_prices, x_capacity):
    item_count = x_weights.shape[0]
    picks_space = 2 ** item_count
    best_price = -1
    best_picks = np.zeros(item_count)
    for p in range(picks_space):
        picks = [int(c) for c in f"{p:0{item_count}b}"]
        price = np.dot(x_prices, picks)
        weight = np.dot(x_weights, picks)
        if weight <= x_capacity and price > best_price:
            best_price = price
            best_picks = picks
    return best_picks


def create_knapsack(item_count=5):
    x_weights = np.random.uniform(1,45, item_count)
    x_prices = np.random.uniform(1,45, item_count)
    x_capacity = np.random.uniform(1,99)
    y = brute_force_knapsack(x_weights, x_prices, x_capacity)
    return x_weights, x_prices, x_capacity, y
```

```python
def knapsack_new_loss(weights,prices,cap,alpha):
    def loss(y_true,y_pred):
        #y_pred = K.round(y_pred)
        #return K.mean(tf.square(K.batch_dot(y_pred, prices, 1) - K.batch_dot(y_true, prices, 1))) + (alpha*(K.mean(K.maximum(((K.batch_dot(y_pred,weights,1)) - cap),0))))
        return K.mean((-1*(K.batch_dot(y_pred, prices, 1))) + (alpha*((K.maximum(K.batch_dot(weights,(y_pred),1) - cap,0)))))
    
    return loss
```

```python
def metric_overprice(prices):
    def overpricing(y_true, y_pred):
        y_pred = K.round(y_pred)
        return K.mean(K.batch_dot(y_pred, prices, 1) - K.batch_dot(y_true, prices, 1))

    return overpricing
```

```python
def metric_space_violation(weights,cap):
    def space_violation(y_true, y_pred):
        y_pred = K.round(y_pred)
        return K.mean(K.maximum(K.batch_dot(y_pred, weights, 1) - cap, 0))

    return space_violation
```
